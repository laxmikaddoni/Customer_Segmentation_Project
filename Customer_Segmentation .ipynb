```python
df['FrequencyOfPurchases'] = df['FrequencyOfPurchases'].map({
    'Weekly': 4,
    'Fortnightly': 2,
    'Monthly': 1
})

df['FrequencyOfPurchases'].fillna(1, inplace=True)
df[['FrequencyOfPurchases']].head()
```

```python
X = df[['Age',
        'PurchaseAmountUSD',
        'PreviousPurchases',
        'FrequencyOfPurchases']]

X.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
X_scaled_df.head()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

df[['Cluster']].head()
```

```python
df['Cluster'].value_counts()
```

```python
df.groupby('Cluster')[[
    'Age',
    'PurchaseAmountUSD',
    'PreviousPurchases',
    'FrequencyOfPurchases'
]].mean()
```

```python
plt.figure()
plt.scatter(
    df['PreviousPurchases'],
    df['PurchaseAmountUSD'],
    c=df['Cluster']
)
plt.xlabel("PreviousPurchases")
plt.ylabel("PurchaseAmountUSD")
plt.title("Customer Segmentation using K-Means")
plt.show()
```
